```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
COLUMNS = ['class', 'age','menopause','tumor-size',
           'inv-nodes','node-caps','deg-malig','breast','breast-quad',
           'irradiat']
# read and store our data
df = pd.read_csv('breast_cancer.data', names=COLUMNS)
# print some item of dataframe
df.head()
```

```python
df.drop(['menopause', 'inv-nodes', 'breast'], axis=1, inplace=True)
df["class"].fillna("no-recurrence-events", inplace=True)
df["age"].fillna("40-49", inplace=True)
df["tumor-size"].fillna("20-24", inplace=True)
df["node-caps"].fillna("no", inplace=True)
df["deg-malig"].fillna("2", inplace=True)
df["breast-quad"].fillna("left_low", inplace=True)
df["irradiat"].fillna("no", inplace=True)
```

```python
df = df.dropna(axis=0)

df[263]
```

```python
df.head()
```

```python
class_maps ={
    "no-recurrence-events":0,
    "recurrence-events":1
}
age_maps ={
    '10-19':0,
    '20-29':1,
    '30-39':2,
    '40-49':3,
    '50-59':4,
    '60-69':5,
    '70-79':6,
    '80-89':7,
    '90-99':8
}
tumor_maps ={
    '0-4':0,
    '5-9':1,
    '10-14':2,
    '15-19':3,
    '20-24':4,
    '25-29':5,
    '30-34':6,
    '35-39':7,
    '40-44':8,
    '45-49':9,
    '50-54':11,
    '55-59':12
}
node_maps = {
    "yes":0,
    "no":1
}

quad_maps = {
    'left_up':0,
    'left_low':1,
    'right_up':2,
    "right_low":3,
    "central":4
}

irradiat_maps ={
    "yes":0,
    "no":1
}
```

```python
df["class"] = df["class"].map(class_maps)
df["age"] = df["age"].map(age_maps)
df["tumor-size"] = df["tumor-size"].map(tumor_maps)
df["node-caps"] = df["node-caps"].map(node_maps)
df["breast-quad"] = df["breast-quad"].map(quad_maps)
df["irradiat"] = df["irradiat"].map(irradiat_maps)
```

```python
features = df.drop('irradiat', axis=1)
features.isnull()
```

```python
labels = df['irradiat']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=60)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor = LogisticRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
prediction = regressor.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(prediction, y_test)
```
